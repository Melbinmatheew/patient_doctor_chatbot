##  Patient - Doctor Dataset sampling

```python
from datasets import load_dataset

# Define the transformation function
def transform_conversation(example):
    # Extract the relevant fields (correcting the case of the column names)
    description = example['Description']
    patient_text = example['Patient'].strip()
    doctor_text = example['Doctor'].strip()

    # Apply the new template for conversation formatting
    reformatted_text = f'<s>[INST] {description} [/INST] {doctor_text} </s>'

    return {'text': reformatted_text}

# Load the dataset
dataset_name = 'Sohaibsoussi/patient_doctor_chatbot'

dataset = load_dataset(dataset_name)

# Shuffle the dataset and slice it
dataset = dataset['train'].shuffle(seed=42).select(range(1000))

# Apply the transformation
transformed_dataset = dataset.map(transform_conversation)

# Select only the 'text' column and remove others
final_dataset = transformed_dataset.remove_columns([col for col in transformed_dataset.column_names if col != 'text'])

# Show a few examples of the final transformed dataset
final_dataset.to_pandas().head()
```

```python
final_dataset.push_to_hub("patient_doctor-1k")
```

### Casual - Conversation Dataset Making

```python
from datasets import load_dataset

# Define the transformation function
def transform_conversation(example):
    # Ensure the fields exist and strip whitespace
    question = example.get('question', '').strip()
    answer = example.get('answer', '').strip()

    # Apply the new template for conversation formatting
    reformatted_text = f'<s>[INST] {question} [/INST] {answer} </s>'
    return {'text': reformatted_text}

# Load the dataset
dataset_name = 'SohamGhadge/casual-conversation'
dataset = load_dataset(dataset_name)

# Shuffle the dataset and slice it
dataset = dataset['train'].shuffle(seed=42).select(range(1000))

# Apply the transformation
transformed_dataset = dataset.map(transform_conversation)

# Show a few examples of the transformed dataset
print(transformed_dataset.to_pandas().head())
```

```python
transformed_dataset.push_to_hub("casual_converations")
```
